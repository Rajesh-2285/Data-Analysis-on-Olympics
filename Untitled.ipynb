```python
sns.countplot(data =df,x = "Sex",palette = 'Set2')
plt.title('Distribution by Gender')
plt.show()
```

```python
sns.histplot(data = df,x = 'Age',bins = 20,kde = True)
plt.xlabel("Age Distribution")
plt.show()
```

```python
sns.histplot(data = df , x='Height',bins = 20,kde = True)
plt.title('Height Distribution')
plt.show()
```

```python
sns.countplot(data = df,x = 'Medal',palette = 'Set2')
plt.title('Medals Distribution')
plt.show()
```

```python
sns.countplot(data = df,x = 'Year',hue ='Medal')
plt.title('Medal Distribution over the Years')
plt.xticks(rotation = 75)
plt.show()
```

```python
sport_event_count = df.groupby('Sport')['Event'].nunique().sort_values(ascending = False)
```

```python
sport_event_count.plot(kind = 'bar')
plt.xlabel('Number of unique events')
plt.ylabel('sport')
plt.title('Number of unique events per sport')
plt.show()
```

```python
year_participants_count = df.groupby('Year')['ID'].nunique()
```

```python
year_participants_count.plot(kind = 'line')
plt.xlabel('Year')
plt.ylabel('Number of participants')
plt.title('Number of participants per year')
plt.show()
```

```python
year_participants_count.plot(kind = 'bar')
plt.xlabel('Year')
plt.ylabel('Number of participants')
plt.title('Number of participants per year')
plt.show()
```

```python
country_avg_age = df.groupby('NOC')['Age'].mean()
```

```python
country_avg_age.sort_values(ascending=False).head(30).plot(kind='bar')
plt.xlabel('country')
plt.ylabel('Average Age')
plt.title('Top countries with Highest average Age by participant')
plt.show()
```

```python
sns.boxplot(data=df,x='Season',y='Age',palette='Set2')
plt.title('Distribution of Ages by Season')
plt.show()
```

```python
sns.violinplot(data =df,x='Medal',y='Height',palette = 'Set2')
plt.xlabel('Medals')
plt.ylabel('Height')
plt.title('Distribution of Heights by Medals')
plt.show()
```

```python
most_medals_country = df.groupby('NOC')['Medal'].value_counts()
most_medals_country.sort_values(ascending = False)
```

```python
medals_country_year = df.pivot_table(index = 'NOC',columns='Year',values='Medal',aggfunc = 'count')
```

```python
medals_country_year.count()
```

```python
sns.heatmap(medals_country_year,linewidths = 0.5)
plt.title('medal counts by country and year')
plt.show()
```
